```python
# DATA PREPROCESSING STARTING.....
# as the value upper age 100 is one(383), this value is going to be dropped
#because it may influence model biasness
clean_dataset = dataset.drop(52)
```

```python
clean_dataset.isnull().sum() #checking null values
```

```python
# the null value of age is filled up by median values
median_age= (clean_dataset['age']).median() #calculating median
clean_dataset['age'] = clean_dataset['age'].fillna(median_age) #filling null with median vlaues
print("the median age is :", median_age)
```

```python
clean_dataset.isnull().sum() #checking null values
```

```python
# Checking unique values in Categorical features
clean_dataset['gender'].value_counts()
```

```python
clean_dataset['ethnicity'].value_counts()
```

```python
# as ethnicity has invalid ( ? ) value, it is replaced by 'Others'
clean_dataset['ethnicity'] = clean_dataset['ethnicity'].replace('?', 'Others')
```

```python
clean_dataset['ethnicity'].value_counts()
```

```python
# as ethnicity has two type others value, it is replaced by 'Others'
clean_dataset['ethnicity'] = clean_dataset['ethnicity'].replace('others', 'Others')
```

```python
clean_dataset['ethnicity'].value_counts()
```

```python
clean_dataset['relation'].value_counts()
```

```python
# as relation has invalid ( ? ) value, it is replaced by 'others'
clean_dataset['relation'] = clean_dataset['relation'].replace('?', 'Others')
```

```python
clean_dataset['relation'].value_counts()
```

```python
# dropping age_description column, as it provides no useful information
clean_dataset = clean_dataset.drop('age_desc', axis=1)
```

```python
clean_dataset.head(5)
```

```python
# dropping result column, it may be produce redundancy as input
clean_dataset = clean_dataset.drop('result', axis=1)
```

```python
clean_dataset.head(5)
```

```python
#Binary Encoding: those colums which has two values ( yes/ no) , converted into 0 and 1

from sklearn.preprocessing import LabelEncoder

binary_columns = ['gender','jundice','austim','used_app_before','Class/ASD']

LE = LabelEncoder()

for col in binary_columns:
    encoded_values = LE.fit_transform(clean_dataset[col])
    clean_dataset[col] = encoded_values
```

```python
clean_dataset.head(5)
```

```python
# one-hot encoding to convert multiple catagories into binary
clean_dataset = pd.get_dummies(clean_dataset, columns=['ethnicity','contry_of_res','relation'], dtype=int)
```

```python
clean_dataset.head(5)
```

```python
#split features and target for input and output #Separate Input Features (X) and Target (y)
#x=dataset[['a','b','c','d']].values #input
#y=dataset[['e']].values   #output
# as there are multiple features, so I am not defining them as parameters, rather I am calling function
x = clean_dataset.drop('Class/ASD', axis='columns')
y = clean_dataset['Class/ASD']
```

```python
#Train/Test method to split the data set into two sets: a training set and a testing set.(80,20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
# features dependency for distance calculations and gradient updates
# age value is large, so if not scaled, the model will be largely biased to age feature
#feature scaling : to scale data all the datas into new values to adjust features that are easier to compare
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scaled = StandardScaler()
x_train = scaled.fit_transform(x_train) # training data
x_test = scaled.transform(x_test) # testing data
```

```python
print(x_train)
```

```python
print(x_train.shape, x_test.shape)
```

```python
# Logistic regression model creation
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
log_reg = LogisticRegression() #default initialization
log_reg = log_reg.fit(x_train,y_train) # training the LR model
y_pred_log = log_reg.predict(x_test) #Predict the response for test dataset
# Model Accuracy, how often is the classifier correct?
accu_log = accuracy_score(y_test, y_pred_log)
print("Logistic regression Classification Accuracy:", accu_log)
```

```python
# support vector machine model creation
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm= SVC(kernel='linear', C=10, random_state=200)
svm=svm.fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
# Model Accuracy, how often is the classifier correct?
accu_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accu_svm)
```

```python
 # ANN creation
## input layer → hidden layer → output layer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
#ann= MLPClassifier(max_iter=1000)
ann = MLPClassifier( hidden_layer_sizes=(16,), activation='relu', solver='adam', max_iter=1000, random_state=42)
ann=ann.fit(x_train, y_train)
y_pred_ann = ann.predict(x_test)
# Model Accuracy, how often is the classifier correct?
accu_ann = accuracy_score(y_test, y_pred_ann)
print("ANN model Accuracy:", accu_ann)
```

```python
# confusion matrix
#how many predictions were correct/incorrect (TP,TN,FP,FN)
from sklearn.metrics import confusion_matrix
print("confusion matrix of Logistic regression model:\n",confusion_matrix(y_test, y_pred_log))
print("confusion matrix of SVM model:\n",confusion_matrix(y_test, y_pred_svm))
print("confusion matrix of ANN model:\n",confusion_matrix(y_test, y_pred_ann))
```

```python
#summarize of confusion matrix
from sklearn.metrics import classification_report
print("Classification report of Logistic Regression:\n",classification_report(y_test, y_pred_log))
print("Classification report of SVM:\n",classification_report(y_test, y_pred_svm))
print("Classification report of ANN:\n",classification_report(y_test, y_pred_ann))
```

```python
# which features are most important for prediction and most important contribution
importance = pd.Series(log_reg.coef_[0], index=x.columns)
importance = importance.sort_values(ascending=False)

print(importance.head(10))
```

```python
# which features are most important for prediction and most important contribution
importance = pd.Series(log_reg.coef_[0], index=x.columns)
# selecting only the questionairies scores
ques_features = importance[['A1_Score','A2_Score','A3_Score','A4_Score','A5_Score',
                         'A6_Score','A7_Score','A8_Score','A9_Score','A10_Score']]
ques_features=ques_features.sort_values(ascending=False)
ques_features.plot.bar(figsize=(20,5), fontsize=14)
plt.title("Feature Importance for ASD Screening Questions (A1–A10)", fontsize=20)
plt.ylabel("Absolute coefficient value", fontsize=20)
plt.show()
```
